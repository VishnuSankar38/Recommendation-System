```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
anime_df = pd.read_csv('/content/drive/MyDrive/Anime.csv')
```

```python
anime_df.head()
```

```python
anime_df['Tags'] = anime_df['Tags'].fillna('')
anime_df['Description'] = anime_df['Description'].fillna('')
anime_df['Tags_Description'] = anime_df['Tags'] + ' ' + anime_df['Description']
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(anime_df['Tags_Description'])
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
def get_recommendations(name, cosine_sim=cosine_sim):
    idx = anime_df.index[anime_df['Name'] == name].tolist()[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    anime_indices = [i[0] for i in sim_scores]
    return anime_df['Name'].iloc[anime_indices]
```

```python
def get_genre_recommendations(genre, anime_df=anime_df):
    genre_anime = anime_df[anime_df['Tags'].str.contains(genre, case=False, na=False)]
    return genre_anime['Name'].head(10)
```

```python
def get_high_rating_recommendations(anime_df=anime_df, n=10):
    high_rating_anime = anime_df.sort_values(by='Rating', ascending=False).head(n)
    return high_rating_anime['Name']
```

```python
print("Content-Based Recommendations:")
print(get_recommendations('Demon Slayer: Kimetsu no Yaiba - Entertainment District Arc'))
```

```python
print("\nGenre-Based Recommendations:")
print(get_genre_recommendations('Action'))
```

```python
print("\nHigh Rating Recommendations:")
print(get_high_rating_recommendations())
```
